```python
# Question 6
df_mar=pd.read_csv("us-marriages-divorces-1867-2014.csv")
mar=[]
div=[]
# creating arrays
for i in range(len(df_mar)):
    a=df_mar["Marriages"][i]/df_mar["Population"][i]
    b=df_mar["Divorces"][i]/df_mar["Population"][i]
    mar.append(a)
    div.append(b)
# count percentages of mar and div over population and append to the array    

plt.plot(df_mar["Year"],mar,label="Marriages")
plt.plot(df_mar["Year"],div,label="Divorces")

# graphing mar and div 

plt.legend() # adding a legend
# Setting axis titltes and graph title 

plt.xlabel("Year")
plt.ylabel("percentage per capita")
plt.title("Marriages and Divorces Per capita In the U.S")    
```

```python
# Question 9
df_roman=pd.read_csv("roman-emperor-reigns.csv")
# reading csv file 
length_re=[] 
label_em=[] 
# creating arrays for reign and emperiors 
for i in range(len(df_roman)):
    if df_roman["Cause_of_Death"][i]=="Assassinated":
        label_em.append(df_roman["Emperor"][i])
        length_re.append(df_roman["Length_of_Reign"][i])
        # retreiving emperior if cause of death = "Assassinated", then add the name to em and reign to re
plt.figure(figsize=(20,15))
# setting pie chart size and creating it
#plt.pie(length_re, labels = label_em,autopct='%1f%%')
# percentage set to 1 decimal place for better view 
plt.pie(length_re, labels = label_em,autopct='%1.1f%%')

plt.title('Fractions of Assasinated Roman Emeprors')
# Adding title then graphing
plt.show()  
```
